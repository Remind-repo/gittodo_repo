```python
import numpy as np
```

```python
# 학생별 과목 성적 데이터
scores_math = [88, 92, 76, 85, 95, 92, 88, 90, 85, 88]
scores_english = [90, 85, 80, 87, 92, 89, 83, 91, 86, 88]
scores_science = [85, 88, 82, 89, 94, 86, 91, 88, 84, 87]
```

```python
# 성적 ndarray화
mathscore = np.array(scores_math)
engscore = np.array(scores_english)
sciencescore = np.array(scores_science)
```

```python
# 1
values, count = np.unique(mathscore, return_counts=True)
mostoccur = values[np.argmax(count)]

print(f"학생 수학 성적의 평균:{mathscore.mean():.2f}, 중앙값:{np.median(mathscore)}, 최빈값:{mostoccur}")
```

```python
# 2
print(f"학생 수학 성적의 분산:{mathscore.var():.2f}, 표준편차:{mathscore.std():.2f}")
```

```python
# 3
print(f"학생 영어 성적의 평균:{engscore.mean():.2f}, 중앙값:{np.median(engscore)}")
```

```python
# 4
values, count = np.unique(sciencescore, return_counts=True)
maxcount = count[np.argmax(count)]
mostoccur = values[count == maxcount]

print(f"과학 성적의 최빈값들은 {", ".join(mostoccur.astype(str))}")
```

```python
person_mean = (mathscore + engscore + sciencescore) / 3
for i, s in enumerate(person_mean):
    print(f"학생 {i+1:2}의 평균 성적은 {s:.2f}")
print(f"\n전체 학생 평균 성적은 {person_mean.mean():.2f}")
```
